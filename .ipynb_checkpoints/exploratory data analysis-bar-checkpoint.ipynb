```python
import pandas as pd
import matplotlib.pyplot as plt

# Exploratory Data Analysis (EDA) - Bar Chart Most Context
df = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

target_columns = [
    { "col": "occ_month", "title": "Total Crime Per Month Comparison", "limit": None },
    { "col": "occ_date", "title": "Date With Most Crimes Commited", "limit": 12 },
    { "col": "AREA NAME", "title": "Area With Most Crimes Commited", "limit": 12 }
]

for dt in target_columns:
    if dt["limit"] == None:
        top_context = df[dt['col']].value_counts()
    else: 
        top_context = df[dt['col']].value_counts().head(dt["limit"])
        
    # Bar Chart
    plt.figure(figsize=(10, 6))
    ax = top_context.plot(kind='bar')
    # Add label
    for i, value in enumerate(top_context):
        plt.text(i, value + 1, str(value), ha='center', va='bottom')
    plt.title(dt["title"])
    plt.xlabel(title)
    plt.ylabel("Number of Crimes")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Bar Chart Top 7 Months With Most Committed Crimes
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Extract month name
data['month_name'] = data['DATE OCC'].dt.month_name()

# Count crime category by month
month_count = data['month_name'].value_counts().head(7)

# Bar Chart
plt.figure(figsize=(10,6))
ax = month_count.plot(kind='bar')
# Add label
for i, value in enumerate(month_count):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
month_count.plot.bar()
plt.title("Top 7 Months With Most Committed Crimes")
plt.xlabel("Month Name")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Bar Chart Top 7 Month-Year With Most Committed Crimes
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Extract month-year name
data['month_year'] = data['DATE OCC'].dt.strftime('%B-%Y')

# Count crime category by month-year
month_year_count = data['month_year'].value_counts().head(7)

# Bar Chart
plt.figure(figsize=(12,6))
ax = month_year_count.plot(kind='bar')
# Add label
for i, value in enumerate(month_year_count):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Top 7 Month-Year With Most Committed Crimes")
plt.xlabel("Month-Year")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Bar Chart Commited Crimes in A Year Comparison
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Extract year name
data['year'] = data['DATE OCC'].dt.strftime('%Y')

# Count crime category by year
year_count = data['year'].value_counts().head(7)

# Bar Chart
plt.figure(figsize=(12,6))
ax = year_count.plot(kind='bar')
# Add label
for i, value in enumerate(year_count):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Commited Crimes in A Year Comparison")
plt.xlabel("Year")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
