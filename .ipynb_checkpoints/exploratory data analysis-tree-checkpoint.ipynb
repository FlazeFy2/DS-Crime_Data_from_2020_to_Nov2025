```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Exploratory Data Analysis (EDA) - Tree Chart Total Crime Per Area
# Dataset
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Count rows
total_context = data.groupby("AREA NAME").size().reset_index(name='Count')

# Sort 
total_context = total_context.sort_values(by='Count', ascending=False).head(20)

# Escape $ 
total_context["AREA NAME"] = total_context["AREA NAME"].astype(str).str.replace('$', r'\$', regex=False)

# Extract
sizes = total_context['Count']
labels = total_context.apply(lambda x: f"{x["AREA NAME"]}\n({x['Count']})", axis=1)

# Plot Tree Map
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=labels, alpha=0.8)
plt.title(f"Total Crime Per Area", fontsize=14)
plt.axis('off')
plt.show()
```
