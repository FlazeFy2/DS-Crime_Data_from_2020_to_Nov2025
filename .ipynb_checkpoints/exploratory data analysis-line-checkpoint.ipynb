```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Crime per Year
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Group by year and count the number of crimes
missions_per_year = data['occ_year'].value_counts().sort_index()

# Plot Line Chart
plt.figure(figsize=(12, 6))
plt.plot(missions_per_year.index, missions_per_year.values, marker='o')
plt.title('Total Crimes Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.grid(True)

# Add Label
for x, y in zip(missions_per_year.index, missions_per_year.values):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=8)
    
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Crime per Month
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Define correct month order
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Group by month and count the number of crimes
missions_per_month = data['occ_month'].value_counts()

# Re-order
missions_per_month = missions_per_month.reindex(month_order)

# Plot Line Chart
plt.figure(figsize=(12, 6))
plt.plot(missions_per_month.index, missions_per_month.values, marker='o')
plt.title('Total Crimes Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.grid(True)

# Add Label
for x, y in zip(missions_per_month.index, missions_per_month.values):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=8)
    
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Crime per Hour (24h format)
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert TIME OCC column to datetime
data['TIME OCC'] = pd.to_datetime(data['TIME OCC'], errors='coerce')

# Extract hour (24h format) from TIME OCC
data['hour'] = data['TIME OCC'].dt.hour

# Group by hour and count the number of crimes
missions_per_hour = data['hour'].value_counts().sort_index()

# Plot Line Chart
plt.figure(figsize=(12, 6))
plt.plot(missions_per_hour.index, missions_per_hour.values, marker='o')
plt.title('Total Crimes Per Hour (24h)')
plt.xlabel('Hour (0-23)')
plt.ylabel('Number of Crimes')
plt.grid(True)

# Add Label
for x, y in zip(missions_per_hour.index, missions_per_hour.values):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=8)
    
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Crime per Quartal
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert Month to Quarter
def quartal_category(month):
    if month in ["Jan", "Feb", "Mar"]:
        return "Q1"
    elif month in ["Apr", "May", "Jun"]:
        return "Q2"
    elif month in ["Jul", "Aug", "Sep"]:
        return "Q3"
    elif month in ["Oct", "Nov", "Dec"]:
        return "Q4"
    else:
        return None

data["occ_quartal"] = data["occ_month"].apply(quartal_category)

# Define correct quartal order
quartal_order = ["Q1","Q2","Q3","Q4"]

# Group by quartal and count the number of crimes
missions_per_quartal = data['occ_quartal'].value_counts()

# Re-order
missions_per_quartal = missions_per_quartal.reindex(quartal_order)

# Plot Line Chart
plt.figure(figsize=(12, 6))
plt.plot(missions_per_quartal.index, missions_per_quartal.values, marker='o')
plt.title('Total Crimes Per Quartal')
plt.xlabel('Quartal')
plt.ylabel('Number of Crimes')
plt.grid(True)

# Add Label
for x, y in zip(missions_per_quartal.index, missions_per_quartal.values):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=8)
    
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Crime per Year-Quartal
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE_OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Create occ_year_quartal in format YYYY-Qn (e.g., 2025-Q1)
data['occ_year_quartal'] = data['DATE OCC'].dt.year.astype(str) + "-Q" + data['DATE OCC'].dt.quarter.astype(str)

# Group by and count
missions_per_quartal = data['occ_year_quartal'].value_counts().sort_index()

# Plot Line Chart
plt.figure(figsize=(12, 6))
plt.plot(missions_per_quartal.index, missions_per_quartal.values, marker='o')
plt.title('Total Crimes Per Year-Quartal')
plt.xlabel('Year-Quartal')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.xticks(rotation=45)
# Add Label
for x, y in zip(missions_per_quartal.index, missions_per_quartal.values):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=8)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Crime per Year-Semester
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE_OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Create occ_year_semester in format YYYY-Sn (e.g., 2025-S1)
data['semester'] = np.where(data['DATE OCC'].dt.quarter <= 2, 1, 2)
data['occ_year_semester'] = data['DATE OCC'].dt.year.astype(str) + "-S" + data['semester'].astype(str)

# Group by and count
missions_per_semester = data['occ_year_semester'].value_counts().sort_index()

# Plot Line Chart
plt.figure(figsize=(12, 6))
plt.plot(missions_per_semester.index, missions_per_semester.values, marker='o')
plt.title('Total Crimes Per Year-Semester')
plt.xlabel('Year-Semester')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.xticks(rotation=45)
# Add Label
for x, y in zip(missions_per_semester.index, missions_per_semester.values):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=8)
plt.show()
```
