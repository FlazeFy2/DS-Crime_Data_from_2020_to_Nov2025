```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Vict Sex','occ_year']

for dt in target_col:
    title = dt.replace("_"," ").title()
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"{title} Comparison")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
df = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Victim Age Category Comparison
# Convert Category
def categorize_category(age):
    if age == 0 or pd.isna(age):
        return None
    elif age <= 12:
        return "Child (0-12)"
    elif age <= 17:
        return "Teen (13-17)"
    elif age <= 24:
        return "Young Adult (18-24)"
    elif age <= 44:
        return "Adult (25-44)"
    elif age <= 64:
        return "Middle Age (45-64)"
    else:
        return "Senior (65+)"

df["Vict Age Category"] = df["Vict Age"].apply(categorize_category)

# Drop missing categories (0 or NaN)
df = df.dropna(subset=["Vict Age"])

top_context = df["Vict Age Category"].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Victim Age Category Comparison")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Pie Chart Commited Crime Comparison By Time Category
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert TIME OCC column to datetime
data['TIME OCC'] = pd.to_datetime(data['TIME OCC'], errors='coerce')

# Extract hour (24h format) from TIME OCC
data['hour'] = data['TIME OCC'].dt.hour

# Convert Category
def time_category(time):
    if time >= 12:
        return "PM"
    else:
        return "AM"

data["Time Category"] = data["hour"].apply(time_category)

top_context = data["Time Category"].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Commited Crime Comparison By Time Category")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Pie Chart Commited Crime Comparison By Time Category
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert TIME OCC column to datetime
data['TIME OCC'] = pd.to_datetime(data['TIME OCC'], errors='coerce')

# Extract hour (24h format) from TIME OCC
data['hour'] = data['TIME OCC'].dt.hour

# Convert Category (Morning, Afternoon, Evening, Middle of Night)
def time_category(hour):
    if 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 18:
        return "Afternoon"
    elif 18 <= hour < 24:
        return "Evening"
    else:
        return "Middle of Night"

data["Time Category"] = data["hour"].apply(time_category)

# Count categories
top_context = data["Time Category"].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Commited Crime Comparison By Time Category")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Pie Chart Commited Crime Comparison By Semester Category
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Extract quarter  convert to semester (S1, S2)
data['semester'] = np.where(data['DATE OCC'].dt.quarter <= 2, "S1", "S2")

# Count category
top_context = data['semester'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Commited Crime Comparison By Semester Category")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Pie Chart Commited Crime Comparison By Weekend vs Weekday
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Extract day & Convert to weekday / weekend
data['day_type'] = np.where(data['DATE OCC'].dt.dayofweek < 5, "Weekday", "Weekend")

# Count category
top_context = data['day_type'].value_counts()

# Pie Chart
plt.figure(figsize=(8,8))
top_context.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title(f"Commited Crime Comparison By Weekend vs Weekday")
plt.ylabel("")
plt.show()
```
