```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Crime per Year
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Group by year and count the number of crimes
missions_per_year = data['occ_year'].value_counts().sort_index()

# Plot Line Chart
plt.figure(figsize=(12, 6))
plt.plot(missions_per_year.index, missions_per_year.values, marker='o')
plt.title('Total Crimes Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.grid(True)

# Add Label
for x, y in zip(missions_per_year.index, missions_per_year.values):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=8)
    
plt.show()
```

```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Declaration
# Exploratory Data Analysis (EDA) - Line Chart Total Crime per Month
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Define correct month order
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Group by month and count the number of crimes
missions_per_month = data['occ_month'].value_counts()

# Re-order
missions_per_month = missions_per_month.reindex(month_order)

# Plot Line Chart
plt.figure(figsize=(12, 6))
plt.plot(missions_per_month.index, missions_per_month.values, marker='o')
plt.title('Total Crimes Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.grid(True)

# Add Label
for x, y in zip(missions_per_month.index, missions_per_month.values):
    plt.text(x, y + 1, str(y), ha='center', va='bottom', fontsize=8)
    
plt.show()
```
