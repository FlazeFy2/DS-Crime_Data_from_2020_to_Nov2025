```python
import pandas as pd
import statistics
import numpy as np

# Declaration
# Descriptive Statistic : Total Crimes and Average Crimes Per Year for Every Area
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Create Year Column
if 'Year' not in data.columns:
    data['Year'] = pd.to_datetime(data['DATE OCC'], errors='coerce').dt.year

# Unique values for Area Name
areas_groups = data['AREA NAME'].unique().tolist()

results = []

for areas in areas_groups:
    # Filter data by AREA NAME
    areas_data = data[data['AREA NAME'] == areas]

    # Total Crimes
    total_crimes = areas_data['DR_NO'].nunique()

    # Average Crimes Per Year
    avg_per_year = round(areas_data.groupby('Year')['DR_NO'].nunique().mean(), 2)

    # Result row
    result_row = {
        'Area Name': areas,
        'Total Crimes': total_crimes,
        'Average Crimes Per Year': avg_per_year
    }

    results.append(result_row)

# Sort
res_df = pd.DataFrame(results)
res_df = res_df.sort_values(by='Average Crimes Per Year', ascending=False)
print(res_df)
```
