```python
import pandas as pd
import matplotlib.pyplot as plt

# Exploratory Data Analysis (EDA) - Bar Chart Most Context
df = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

target_columns = [
    { "col": "occ_month", "title": "Total Crime Per Month Comparison", "limit": None },
    { "col": "occ_date", "title": "Date With Most Crimes Commited", "limit": 12 },
    { "col": "AREA NAME", "title": "Area With Most Crimes Commited", "limit": 12 }
]

for dt in target_columns:
    if dt["limit"] == None:
        top_context = df[dt['col']].value_counts()
    else: 
        top_context = df[dt['col']].value_counts().head(dt["limit"])
        
    # Bar Chart
    plt.figure(figsize=(10, 6))
    ax = top_context.plot(kind='bar')
    # Add label
    for i, value in enumerate(top_context):
        plt.text(i, value + 1, str(value), ha='center', va='bottom')
    plt.title(dt["title"])
    plt.xlabel(title)
    plt.ylabel("Number of Crimes")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Bar Chart Top 7 Months With Most Committed Crimes
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Extract month name
data['month_name'] = data['DATE OCC'].dt.month_name()

# Count crime category by month
month_count = data['month_name'].value_counts().head(7)

# Bar Chart
plt.figure(figsize=(10,6))
ax = month_count.plot(kind='bar')
# Add label
for i, value in enumerate(month_count):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
month_count.plot.bar()
plt.title("Top 7 Months With Most Committed Crimes")
plt.xlabel("Month Name")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Bar Chart Top 7 Semester-Year With Most Committed Crimes
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Extract semester-year
data['semester_year'] = data['DATE OCC'].dt.to_period('2Q').astype(str).apply(
    lambda x: f"S{1 if x.split('Q')[1].startswith('1') else 2}-{x[:4]}"
)

# Count crime by semester-year
semester_year_count = data['semester_year'].value_counts().head(7)

# Bar Chart
plt.figure(figsize=(12,6))
ax = semester_year_count.plot(kind='bar')

# Add label
for i, value in enumerate(semester_year_count):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.title("Top 7 Semester-Year With Most Committed Crimes")
plt.xlabel("Semester-Year")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Declaration
# Exploratory Data Analysis (EDA) - Bar Chart Top 7 Week-Month-Year With Most Committed Crimes
data = pd.read_csv("Crime_Data_from_2020_to_Present.csv")

# Convert DATE OCC column to datetime
data['DATE OCC'] = pd.to_datetime(data['DATE OCC'], errors='coerce')

# Extract week-month-year
def get_custom_week(day):
    if day <= 7:
        return 1
    elif day <= 14:
        return 2
    elif day <= 21:
        return 3
    elif day <= 28:
        return 4
    else:
        return 5

data['week'] = data['DATE OCC'].dt.day.apply(get_custom_week)
data['month'] = data['DATE OCC'].dt.strftime('%b')
data['year'] = data['DATE OCC'].dt.year

data['week_month_year'] = data.apply(
    lambda x: f"Week-{x['week']} {x['month']} {x['year']}",
    axis=1
)

# Count crime category by week-month-year
week_month_year_count = data['week_month_year'].value_counts().head(7)

# Bar Chart
plt.figure(figsize=(12,6))
ax = week_month_year_count.plot(kind='bar')
# Add label
for i, value in enumerate(week_month_year_count):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')
plt.title("Top 7 Week-Month-Year With Most Committed Crimes")
plt.xlabel("Week-Month-Year")
plt.ylabel("Number of Crimes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
